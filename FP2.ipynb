# Data Preprocessing
1. Feature Engineering

```python
# Creating a new data frame containing mean value of annual rainfall in different subdivisions

x=pd.DataFrame(df.groupby("SUBDIVISION")['ANNUAL'].mean())
x
```

```python
x.shape
```

```python
x=x.reset_index() 
x
```

```python
# Creating a new dataframe by merging df and x 

df1 = df.merge(x, on='SUBDIVISION', how='left')
```

```python
df1.head()
```

```python
# Assigning mean value of annual rainfall to dataset as a new column

df['MEAN ANNUAL']=df1['ANNUAL_y']
```

```python
df.head()
```

#Creating a new column as Prediction which predicts 'flood', if annual rainfall is 25% greater than mean of annual rainfall of that subdivision, if 25% lesser than the the mean of annual rainfall of that subdivision, predict "drought"  if otherwise 'normal'



```python
#Creating a new column as Prediction which predicts flood as yes(1) if sum of monthly rainfall is greater than mean of annual rainfall of that subdivision and no(0) if otherwise

df['Prediction1'] = np.where(df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].sum(axis=1)>df['MEAN ANNUAL']*1.25, 1, 0)
df
```

```python
#Creating a new column as Prediction2 which predicts drought as yes(2) if sum of monthly rainfall is less than mean of annual rainfall of that subdivision and no(0) if otherwise

df['Prediction2'] = np.where(df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].sum(axis=1)<df['MEAN ANNUAL']*0.75, 2, 0)
df
```

```python
df['Prediction'] = df['Prediction1'] + df['Prediction2']
df
```

```python
df.drop(['Prediction1','Prediction2'], axis = 1)
```
